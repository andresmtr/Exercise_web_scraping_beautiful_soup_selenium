```python

### Library to web scraping
### Librerias para web scraping
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By


### Library to clean data base
### Librerias para limpiar base
import pandas as pd
```

```python
### Used web driver google chrome / important downloand
### Uso del motor de busqueda de Google Chrome / importante descargar


path_drive = "/Users/andresmauriciotrianareina/Documents/Desarrollos/21_scraping_paginas_amarillas/chromedriver"

driver = webdriver.Chrome(executable_path=path_drive)
```

```python
### Verify if page working
### Verificar que la pagina funcione

link_web = 'https://amarillas.emol.com/abogados/todo-chile?p=1'

website = requests.get(link_web)

website
```

```python
### List to save de information
### listas para guardar la información

urls_list = []
url_pagina = []
name_list = []
direccion_list = []
website_list = []
categoria_list = []
telefono_list = []
```

```python

### Open subpages
### Abrir cada sub pagina

for i in range(2): 

    driver.get('https://amarillas.emol.com/abogados/todo-chile?p='+str(i+1))
    soup = BeautifulSoup(driver.page_source, 'lxml')
    paginas = soup.findAll('a', attrs = {'class':'tmb', 'ng-click':'$event.preventDefault(); getProfile(profile_url, result._source.anunciador, result._source.meco_code, priority, result._source.direcciones[0].comuna, result._source.direcciones[0].region, ($index+1), StatsData.STATS_CODIGO_MARCA_LISTADO_CENTRAL_CLICK_LOGO)'})
    link_web = 'https://amarillas.emol.com'

    for link in paginas:
        urls_list.append(link_web+link.get('ng-href'))
        url_pagina.append(link.get('ng-href'))
    
```

```python
### Dataframe with links 
### Dataframe con enlaces

df = pd.DataFrame(list(zip(urls_list, url_pagina)), columns=['Profile Páginas amarillas', 'partial url'])
df.head()
```

```python
### Cicle For to extract data to specific xpath
### Ciclo For para extraer los datos con el xpath


for i in range(len(df.index)):
    driver.get(df['Profile Páginas amarillas'][i])

    try:

        name = driver.find_element(By.XPATH, '//*[@id="search"]/article/div[1]/h1').text
        name_list.append(name)

        telefono = driver.find_element(By.XPATH, './/*[@id="items"]/span/li/span/span').text
        telefono_list.append(telefono)

        direccion = driver.find_element(By.XPATH, './/*[@id="items"]/span/ul/a/p').text
        direccion_list.append(direccion)

        etiquetas = driver.find_element(By.XPATH, './/*[@id="search"]/article/h2').text
        categoria_list.append(etiquetas)
    
    except:
        
        #name_list.append('')
        telefono_list.append('')
        direccion_list.append('')
        categoria_list.append('')
```

```python
### Add to Dataframe the information
### Agregar al datafrema la información

df['Nombre organización'] = name_list
df['Telefono'] = telefono_list
df['Direccion'] = direccion_list
df['Categorias'] = categoria_list
```

```python
### Export the dataframe to csv
### Exportar el dataframe a csv

df.to_csv('base.csv')
```
